### 2) Data Cleaning: Checking for Null Values.
* Before any data exploration can be carried out, the data must be cleaned of any null or missing entries.
* From this we can see that:
   >* *`outcome_subtype`* has **43,324** null values
   >* *`outcome_type`* has **10** null values
   >* *`sex_upon_outcome`* has **1** null value
   >* *`sex_upon_intake`* has **1** null value
* The data also contains several columns which contain either duplicate or unecessary data for this EDA, so these will also be removed.

```python
# Setting up a function names null_values(), which will output a table counting the nulls and the percentage it makes up from the overall data.
def null_values(dataframe):
    # Using .isnull.sum(), to count the number of null values present across all columns.
    null_values = dataframe.isnull().sum() 

    # Counting the total number of entries in the dataframe.
    total_cnt = len(dataframe) 

    # Creates a new dataframe in which the number of null values is the first column.
    null_values = pd.DataFrame(null_values,columns = ['Null Count'])

    # This line returns what percentage of the data is null for each column.
    null_values['Percentage (%)'] = round((null_values['Null Count']/total_cnt)*100, 3)

    # This line returns a dataframe which is ordered by the percentage of nulls, from highest to lowest.
    return null_values.sort_values('Percentage (%)', ascending=False)


# Calling the defined function above taking the TMDB as its parameter.
null_values(df_eda)
```

### 3) Data Cleaning: Dealing with Null Values.

* Now that the null values have been identified, they can be dealt with.
* Outline of null value cleaning:
    >* There are several columns that are not required for the following EDA and can be dropped before continuing with this exploration.
    >* The **`outcome_subtype` and `outcome_type`** columns can be filled with a flag, to account for the missing values.
    >* The **`sex_upon_outcome` and `sex_upon_intake`** column both contain a single entry with a null value - this comes from the **same observation**\
       so this single observation can be dropped from the dataframe to account for the null values.
  

```python
# Dropping columns that will be irrelevant for the EDA.
df_eda.drop(columns = ['age_upon_outcome', 'age_upon_intake', 'dob_year', 'dob_month',
            'dob_monthyear', 'outcome_month', 'outcome_year', 'outcome_monthyear',
            'outcome_weekday', 'outcome_hour', 'intake_month', 'intake_year',
            'intake_monthyear', 'intake_weekday', 'intake_hour', 'count', 
            'outcome_number', 'intake_number', 'time_in_shelter', 'animal_id_intake'],
           inplace = True)
```

```python
# Filling the missing values in 'outcome_subtype' with a string.
df_eda['outcome_subtype'].fillna('No subtype', inplace = True)

#Filling in the missing values in 'outcome_type' with a string.
df_eda['outcome_type'].fillna('No type', inplace = True)

# Dropping the single entry with a missing value in 'sex_upon_outcome' - this also deals with the entry missing in 'sex_upon_intake'.
df_eda.dropna(subset = ['sex_upon_outcome'], inplace = True)
```

```python
# Checking the null values again to ensure they have been dealt with.
null_values(df_eda)
```

##### **From the above dataframe, we can see no null values remain in the data. The data is suitable to be used further.**

#### Before continuing, I made the following changes to the columns.
>* Separated the spayed/neutered status from the sex of the animal and created two new columns: `is_spayed_neutered` and `animal_sex`.
>* Created a new column called `is_adopted` to contain the adoption status of the animal.

```python
# Creates a new column which separates the animals spayed/neutered status from its sex.
df_eda['is_spayed_neutered'] = df_eda['sex_upon_outcome'].str.contains('Spayed|Neutered')

# Creates a new column containing just the animal's sex - using indexing as the sex comes last in the 'sex_upon_outcome' column.
df_eda['animal_sex'] = df_eda['sex_upon_outcome'].str.split().str[-1]

# Creates a new column which contains the animal's adoption status.
df_eda['is_adopted'] = df_eda['outcome_type'].isin(['Adoption', 'Barn Adoption']).astype(int)
```

#### 4.1) Adoption Success by Animal Type
* Which type of animal is most likely to be adopted?

```python
# Setting figure size.
plt.figure(figsize = (6, 6))

# Creating a barplot using the animal_type as the data.
sns.countplot(data=df_eda, x='animal_type', hue='is_adopted')

# Setting plot labels.
plt.title('Adoption Success by Animal Type')
plt.xlabel('Animal Type')
plt.ylabel('Number of Animals')
plt.legend(title='Is Adopted?', labels=['Not Adopted', 'Adopted'])
plt.xticks(rotation=45)
plt.show()
```

##### Findings:
* From the above barchart, it is clear that dogs have the highest adoption rate followed by cats.

#### 4.2) Adoption Success by Age Group
* What age category of animal is most likely to be adopted?

```python
# Creating bins to categorise the data into.
bins = [0, 90, 365, 1825, 3650, np.inf]

# Creating labels for each of the bins - to be used later.
labels = ['Kitten/Puppy', 'Junior', 'Adult', 'Senior', 'Elderly']

# Creates a new column called age group and categorises the ages using the labels mentioned above.
df_eda['age_group'] = pd.cut(df_eda['age_upon_outcome_(days)'], bins = bins, labels = labels, right = False)

# Setting the figure size.
plt.figure(figsize = (6, 6))

# Creating a barplot using the age_group data 
sns.countplot(data = df_eda, x = 'age_group', hue = 'is_adopted', order=labels)

# Setting plot labels.
plt.title('Adoption Success by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Number of Animals')
plt.legend(title='Is Adopted?', labels=['Not Adopted', 'Adopted'])
plt.xticks(rotation=45)
plt.show()
```

##### Findings:
* From the above barchart, adult pets tend to be adopted the most, but considering proportions, the kitten/puppy age group has the highest adoption rates.

#### 4.3) Adoption Success by Sex
* Do people want to adopt male or female animals?

```python
# Setting the figure size.
plt.figure(figsize = (6, 6))

# Plotting a barplot using the animal_sex data.
sns.countplot(data = df_eda, x = 'animal_sex', hue = 'is_adopted')

# Setting plot labels.
plt.title('Adoption Success by Sex')
plt.xlabel('Sex')
plt.ylabel('Number of Animals')
plt.legend(title='Is Adopted?', labels=['Not Adopted', 'Adopted'])
plt.show()
```

##### Findings:
* From the above barchart, there seems to be no preference for either sex and both male and female animals are adopted in roughly equal proportions.

#### 4.4) Distribution of Time Spent in the Shelter (Adopted vs Non-adopted Pets)
* How long did animals spend in the shelter? - Comparison between adopted and non-adopted pets

```python
# Setting figure size
plt.figure(figsize = (6, 6))

# Plotting a histogram to show the distribution.
sns.histplot(data = df_eda, x = 'time_in_shelter_days', hue = 'is_adopted', multiple = 'dodge', bins = 50)

# Setting plot labels.
plt.title('Time in Shelter Distribution for Adopted vs. Not Adopted')
plt.xlabel('Time in Shelter (Days)')
plt.ylabel('Number of Animals')
plt.xlim(0, 300)
plt.legend(title='Is Adopted', labels=['Not Adopted', 'Adopted'])
plt.show()
```

##### Findings:
* The above barchart shows that adopted animals are generally in the shelter for a shorter period than those with other outcomes. The distribution for adopted animals is skewed heavily towards a short time in the shelter, suggesting that animals who are adopted, find their homes relatively quickly.

#### 4.5)  Distribution of Animal Types in the Shelter
* What is the distribution of animal types in the shelter?

```python
# Setting figure size.
plt.figure(figsize = (10, 5))

# Creating a variable called intake_counts which holds the count of each animal type.
intake_counts = df['animal_type'].value_counts()

# Plotting a pie chart to show the distribution of each animal type
plt.pie(intake_counts, labels = intake_counts.index, autopct = '%1.1f%%', startangle = 140, colors = sns.color_palette("pastel"))
plt.title('Distribution of Animal Types')
plt.show()
```

##### Findings:
* The above pie chart shows that the majority of the animals present in the shelter are dogs, followed by cats, other types of animals and finally birds.

#### 4.6)  Distribution of Intake Types
* What is the distribution of the types of intake?

```python
# Setting figure size.
plt.figure(figsize = (10, 5))

# Setting a variable called intake_counts which counts each intake_type
intake_counts = df['intake_type'].value_counts()

# Plotting a pie chart.
plt.pie(intake_counts, labels = intake_counts.index, autopct = '%1.1f%%', startangle = 140, colors = sns.color_palette("pastel"))
plt.title('Distribution of Intake Types')
plt.show()
```

##### Findings:
* The above pie chart shows that the majority of the animals were strays when taken in by the shelter.

#### 4.7) Distribution of Intake Condition
* What were the conditions of the animals when they were taken in by the shelter?

```python
# Setting figure size.
plt.figure(figsize = (10, 5))

# Creating a variable called condition_counts to hold the counts for each of the conditions when the animals were taken in.
condition_counts = df['intake_condition'].value_counts()

# Plotting a barplot to show the condition counts for each condition
sns.barplot(x = condition_counts.index, y = condition_counts.values, palette = 'Spectral')

# Setting figure labels.
plt.title('Distribution of Intake Condition')
plt.xlabel('Condition')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

##### Findings:
* From the barplot above, it is clear that the majority of animals taken in by the shelter were 'normal' and in comparison only a small proportion were injured, sick or nursing and a very small proportion of these animals were either aged, feral, pregnant or came in some other condition.


#### 4.8) Correlation Matrix of Numerical Features

```python
# Picking out the relative numerical features.
numeric_df = df_eda.select_dtypes(include = np.number)

# Creating a correlation matrix using the features outlined above.
correlation_matrix = numeric_df.corr()


# Setting figure size.
plt.figure(figsize = (6, 6))

# PLotting using a heatmap.
sns.heatmap(corr_matrix, annot = True, cmap = 'coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Key Features')
```
